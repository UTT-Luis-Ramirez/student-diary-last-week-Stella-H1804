##*2.Medidas de Tendencia central* ❓

Para un conjunto de datos la medida más conocida y útil es la media.

Las medidas de tendencia central son aquellas que en estadística nos ayudan a definir los datos en concreto.

+ Media: es el promedio de todos los datos de la muestra.

+ Mediana: es el valor del medio de todos los datos ordenados de menor a mayor.

+ Moda: es el valor que más se repite del conjunto de datos.

Siendo estos los más conocidos pero también encontramos más tipos, según sean poblacionales o muestrales.

##**DATOS ORDENADOS**

En estadística también nos encontramos el poncentaje que divide un conjunto de datos presentados, los cuales son:

+ **DECILES**: Son la 10 parte de los datos, representa el 10% de 100%

+ **CUARTILES**: Son la cuarta parte de los datos, representa el 25% de 100%

+ **PERCENTILES**: Es la sentena parte del entero de los datos agrupados.

Para calcular la posición de los cuartiles, deciles o percentiles debes el número por la suma del número total de datos más uno y dividir el resultado entre 4, 10 o 100 respectivamente.

##*MEDIDAS DE DISPERSIÓN* ✨

Las medidas de tendencia central buscan resumir un sólo conjunto, pero no describen en qué medida difieren entre sí los valores.

Para este tipo de casos nos encontramos:

+ **VARIANZA**: Es el promedio de las distancias al cuadrado que van desde las observaciones a la media, también tenemos la varianza muestral.

+ **DESVIACIÓN ESTÁNDAR**: Es la raíz cuadrada del promedio de las distancias al cuadrado que van desde las observaciones a la media.

+ **RANGO**: Es el valor númerico que indica la diferencia entre el valor máximo y mínimo o muestra estadística, también tenemos rango medio.

En las fórmulas también nos encontramos el valor absoluto, el cual es el mismo número sin tener en cuenta el valor positivo o negativo.
